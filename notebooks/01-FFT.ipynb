```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
Fs = 1  # Hz
N = 100  # number of points, our FFT size
```

```python
# Create a simple sine signal with a frequency of 0.15 Hz w/ sampling rate of 1 Hz
t = np.arange(N)
s = np.sin(0.15 * 2 * np.pi * t)
# Windowing the signal, taper the first and last values to 0 to reduce the edge effects
s = s * np.hanning(N)  # apply a Hanning window
```

```python
# FFT Sizing should be a power of 2 (Generally between 128 & 4096)

# This will result in an array of complex numbers
S = np.fft.fft(s)

# This will result in a not centered FFT, let's shift it
S_shifted = np.fft.fftshift(S)

# Let's plot the magnitude of the FFT
S_mag = np.abs(S_shifted)

# Let's plot the phase of the FFT
S_phase = np.angle(S_shifted)

f = np.arange(Fs/-2, Fs/2, Fs/N)

plt.figure(0)
plt.plot(f, S_mag, ".-")
plt.xlabel("FFT Index")
plt.ylabel("FFT Of Signal (Magnitude)")
plt.figure(1)
plt.plot(f, S_phase, ".-")
plt.xlabel("FFT Index")
plt.ylabel("FFT Of Signal (Phase)")
plt.show()
```
