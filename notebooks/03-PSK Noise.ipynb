```python
# Imports
import numpy as np
import matplotlib.pyplot as plt
```

```python
num_symbols = 1000
```

```python
x_int = np.random.randint(0, 4, num_symbols)  # 0 to 3
x_degrees = (x_int * 360.0 / 4.0) + 45
x_radians = x_degrees * np.pi / 180.0
x_symbols = np.cos(x_radians) + 1j * np.sin(x_radians)  # Produces complex QPSK symbols
plt.plot(np.real(x_symbols), np.imag(x_symbols), ".")
plt.grid(True)
plt.show()
```

In the figure above, there is **no noise**; the symbols overlap perfectly.

Let's add some noise

```python
# Additive White Gaussian Noise w/ unity power
n = (np.random.randn(num_symbols) + 1j * np.random.randn(num_symbols)) / np.sqrt(2)
noise_power = 1e-2
r = x_symbols + n * np.sqrt(noise_power)
plt.plot(np.real(r), np.imag(r), ".")
plt.grid(True)
plt.show()
```

```python
# Add phase noise (phase jitter in the LO)
phase_noise = np.random.randn(len(x_symbols)) * 0.1  # 0.1 is the strength of the phase noise
r = x_symbols * np.exp(1j * phase_noise)
plt.plot(np.real(r), np.imag(r), ".")
plt.grid(True)
plt.show()
```

```python
# The complete experience
r = (x_symbols + n * np.sqrt(noise_power)) * np.exp(1j * phase_noise)
plt.plot(np.real(r), np.imag(r), ".")
plt.grid(True)
plt.show()
```
